```python
# 5. Handling Missing Data - identify and handle missing values using techniques 
# like imputation or removal.

missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)

df = df.fillna(df.mean(numeric_only=True))
```

```python
# Fill Missing Values
# Fill missing values with the mode (most frequent value) for each column
df = df.fillna(df.mode().iloc[0])

# Display the DataFrame with missing values filled
print(df)
```

```python
# Check for missing values after Fillna
# 5. Handling Missing Data - identify and handle missing values using techniques 
# like imputation or removal.

missing_values_after = df.isnull().sum()
print("\nMissing Values:")
print(missing_values_after)

df = df.fillna(df.mean(numeric_only=True))
```

```python
# 7. Data Visualization - Create visualizations to explore data distributions, relationships, and patterns. 
#Use libraries like Matplotlib and Seaborn.

#a. Univariate plots focus on a single variable or feature at a time. 
#They are used to visualize the distribution and characteristics of individual variables in isolation.

#Histogram 
# Assuming 'Selling_Price' is a column in your DataFrame 'df'
plt.figure(figsize=(8, 6))

# Create a histogram for 'Selling_Price'
sns.histplot(data=df, x='Selling_Price', kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')

plt.show()
```

```python
# Density Plots
# Assuming 'Selling_Price' is a column in your DataFrame 'df'
plt.figure(figsize=(8, 6))

# Create a density plot for 'Selling_Price'
sns.kdeplot(data=df, x='Selling_Price', fill=True)
plt.title('Density Plot of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Density')

plt.show()
```

```python
# Box and Whisker Plots
# Assuming 'Year' is a column in your DataFrame 'df'
plt.figure(figsize=(8, 6))

# Create a box plot for 'Year'
sns.boxplot(data=df, x='Year')
plt.title('Distribution of Car Year')
plt.xlabel('Year')

plt.show()
```

```python
#Multivariate plots involve the simultaneous visualization of two or more variables to explore relationships, 
# correlations, and patterns among them.


#method: {pearson, kendall, spearman}
# Calculate the correlation matrix with numeric_only set to True
correlations = df.corr(method='pearson', numeric_only=True)    #Pearson Method

# Create a mask to hide the upper triangle for better visualization
matrix = np.triu(correlations)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm', mask=matrix)
plt.title('Correlation Matrix (Pearson)')
plt.show()
```

```python
#Multivariate plots involve the simultaneous visualization of two or more variables to explore relationships, 
# correlations, and patterns among them.


#method: {pearson, kendall, spearman}
# Calculate the correlation matrix with numeric_only set to True
correlations = df.corr(method='kendall', numeric_only=True)    #Kendall Method

# Create a mask to hide the upper triangle for better visualization
matrix = np.triu(correlations)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm', mask=matrix)
plt.title('Correlation Matrix (Kendall)')
plt.show()
```

Present_Price: The current market price of the car is positively correlated with the selling price. Cars with higher market prices tend to have higher selling prices.
Driven_kms: The number of kilometers driven by the car is negatively correlated with the selling price. Cars with lower mileage often have higher selling prices.
Fuel_Type: The type of fuel used can affect the selling price. Diesel cars typically have higher selling prices compared to Petrol or CNG cars.
Transmission: Automatic transmission cars tend to have higher selling prices compared to manual transmission cars.
Prev_Owner (Except First Owner): The number of previous owners can influence the selling price. Cars with fewer previous owners tend to have higher selling prices.
